Trong một tập dữ liệu có 99% là lớp A và 1% là lớp B, mô hình có thể đạt được tổng lỗi rất thấp bằng cách luôn dự đoán lớp A. **Câu 4:** Vì P-R Curve nhạy cảm hơn với hiệu năng trên lớp thiểu số. Trục hoành của ROC là FPR (`FP / (FP + TN)`), có mẫu số là tổng số lớp Negative (rất lớn trong dữ liệu mất cân bằng). Một lượng lớn FP có thể chỉ tạo ra một FPR rất nhỏ, làm cho đường cong ROC trông rất tốt. Ngược lại, trục tung của P-R Curve là Precision (`TP / (TP + FP)`), có mẫu số chứa FP. Cùng một lượng lớn FP đó sẽ làm mẫu số của Precision tăng vọt, khiến Precision giảm mạnh. P-R Curve sẽ "vạch trần" hiệu năng kém mà ROC Curve có thể che giấu.

**Câu 5:** `stratify=y` đảm bảo rằng khi chia dữ liệu thành tập train và test, tỷ lệ phân phối của các lớp trong `y` sẽ được giữ nguyên ở cả hai tập. Điều này cực kỳ quan trọng với dữ liệu mất cân bằng. Nếu không có `stratify`, có khả năng (dù nhỏ) là tập test của bạn có thể không chứa mẫu nào của lớp thiểu số, hoặc tỷ lệ trong tập train và test khác xa thực tế, dẫn đến việc huấn luyện và đánh giá mô hình không đáng tin cậy.

#### Đáp án bài tập Code

```python
# Bài tập 1: Lời giải
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Tạo dữ liệu
X, y = make_classification(
    n_samples=2000,
    n_features=5,
    n_informative=2,
    n_redundant=0,
    n_classes=2,
    weights=[0.99, 0.01], # Tỷ lệ 99/1
    random_state=42
)

# 2. Trực quan hóa
plt.figure(figsize=(6, 4))
sns.countplot(x=y)
plt.title('Phân phối các lớp (Tỷ lệ 99/1)')
plt.show()

# 3. Chia dữ liệu
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

# 4. Huấn luyện mô hình
model = LogisticRegression(random_state=42)
model.fit(X_train, y_train)

# 5. In classification report
y_pred = model.predict(X_test)
report = classification_report(y_test, y_pred)
print(report)

# 6. Nhận xét
print("Nhận xét:")
print("- Accuracy tổng thể rất cao (khoảng 99%), gây hiểu lầm.")
print("- Nhìn vào lớp 1 (thiểu số), chỉ số Recall rất thấp (thường là 0.00 hoặc một số rất nhỏ).")
print("- Điều này có nghĩa là mô hình gần như đã bỏ lỡ TOÀN BỘ các mẫu của lớp thiểu số.")
print("- Đây là một mô hình vô dụng trong thực tế để phát hiện lớp 1.")
```

## Tại sao Đánh giá Mô hình lại Quan trọng?

Trong một dự án Data Science, việc xây dựng một mô hình Machine Learning không chỉ dừng lại ở việc `model.fit(X, y)`. Chúng ta cần phải biết mô hình đó hoạt động tốt đến mức nào, có đáng tin cậy để đưa vào ứng dụng thực tế hay không. Quá trình này được gọi là **đánh giá mô hình (model evaluation)**.

Hãy tưởng tượng bạn xây dựng một mô hình dự đoán bệnh ung thư. Nếu mô hình của bạn dự đoán sai, hậu quả có thể rất nghiêm trọng. Một người bệnh có thể bị chẩn đoán là khỏe mạnh và không được điều trị kịp thời. Ngược lại, một người khỏe mạnh có thể bị chẩn đoán sai, dẫn đến lo lắng và các xét nghiệm không cần thiết.

Đánh giá mô hình giúp chúng ta:
- **Đo lường hiệu quả:** Cung cấp những con số cụ thể để đo lường mức độ chính xác của mô hình.
- **So sánh các mô hình:** Giúp lựa chọn ra mô hình tốt nhất trong số nhiều mô hình đã xây dựng.
- **Hiểu rõ hơn về mô hình:** Phân tích các trường hợp mô hình dự đoán đúng/sai, từ đó tìm cách cải thiện.
- **Đảm bảo tính công bằng và đạo đức:** Tránh các sai lệch (bias) không mong muốn, ví dụ như mô hình hoạt động tốt trên nhóm người này nhưng lại kém hiệu quả trên nhóm người khác.

## Các Loại Bài Toán Machine Learning

Trước khi đi sâu vào các chỉ số đánh giá, chúng ta cần phân biệt các loại bài toán chính trong Machine Learning vì mỗi loại sẽ có cách đánh giá khác nhau.

1.  **Supervised Learning (Học có giám sát):** Dữ liệu đầu vào được gán nhãn sẵn. Mô hình học cách ánh xạ từ đầu vào (features) đến đầu ra (labels).
    *   **Classification (Phân loại):** Đầu ra là một danh mục (category). Ví dụ: dự đoán email là spam hay không spam, phân loại ảnh chó hay mèo.
    *   **Regression (Hồi quy):** Đầu ra là một giá trị liên tục. Ví dụ: dự đoán giá nhà, dự báo nhiệt độ.

2.  **Unsupervised Learning (Học không giám sát):** Dữ liệu đầu vào không có nhãn. Mô hình tự tìm ra cấu trúc hoặc các mẫu ẩn trong dữ liệu.
    *   **Clustering (Phân cụm):** Nhóm các điểm dữ liệu tương tự nhau thành các cụm. Ví dụ: phân nhóm khách hàng.
    *   **Dimensionality Reduction (Giảm chiều dữ liệu):** Giảm số lượng biến đầu vào mà vẫn giữ được thông tin quan trọng.

3.  **Reinforcement Learning (Học tăng cường):** Mô hình (agent) học bằng cách tương tác với môi trường và nhận phần thưởng (reward) hoặc phạt (penalty) cho các hành động của mình. Ví dụ: huấn luyện AI chơi game, xe tự lái.

Trong khuôn khổ khóa học này, chúng ta sẽ tập trung chủ yếu vào bài toán **Classification** trong Supervised Learning, vì đây là nơi vấn đề mất cân bằng dữ liệu (imbalanced data) thường xảy ra và có ảnh hưởng lớn nhất.

## Đánh giá Mô hình Phân loại (Classification)

Để đánh giá một mô hình phân loại, chúng ta so sánh kết quả dự đoán của mô hình (`y_pred`) với kết quả thực tế (`y_true`). Bảng tổng hợp kết quả so sánh này được gọi là **Confusion Matrix (Ma trận nhầm lẫn)**.

### Confusion Matrix

Đối với bài toán phân loại nhị phân (binary classification) với hai lớp là **Positive (1)** và **Negative (0)**, confusion matrix sẽ có dạng 2x2:

|                  | **Predicted: Negative (0)** | **Predicted: Positive (1)** |
|------------------|-----------------------------|-----------------------------|
| **Actual: Negative (0)** | True Negative (TN)          | False Positive (FP)         |
| **Actual: Positive (1)** | False Negative (FN)         | True Positive (TP)          |

- **True Positive (TP):** Mô hình dự đoán là Positive, và thực tế đúng là Positive.
  *Ví dụ: Dự đoán bệnh nhân bị bệnh, và thực tế họ bị bệnh.*
- **True Negative (TN):** Mô hình dự đoán là Negative, và thực tế đúng là Negative.
  *Ví dụ: Dự đoán bệnh nhân không bị bệnh, và thực tế họ không bị bệnh.*
- **False Positive (FP):** Mô hình dự đoán là Positive, nhưng thực tế là Negative (Lỗi loại I - Type I Error).
  *Ví dụ: Dự đoán bệnh nhân bị bệnh, nhưng thực tế họ không bị bệnh.*
- **False Negative (FN):** Mô hình dự đoán là Negative, nhưng thực tế là Positive (Lỗi loại II - Type II Error).
  *Ví dụ: Dự đoán bệnh nhân không bị bệnh, nhưng thực tế họ bị bệnh.*

Trong hai loại lỗi, **False Negative (FN)** thường nguy hiểm hơn trong nhiều bài toán thực tế (như y tế, an ninh), vì nó bỏ sót một trường hợp quan trọng cần xử lý.

### Ví dụ với Python

Hãy cùng xem một ví dụ đơn giản để xây dựng confusion matrix.

```python
from sklearn.metrics import confusion_matrix
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Dữ liệu thực tế (ground truth)
y_true = [1, 0, 1, 1, 0, 1, 0, 0, 1, 0]

# Dữ liệu dự đoán bởi mô hình
y_pred = [1, 1, 1, 0, 0, 1, 0, 0, 1, 0]

# Tính confusion matrix
cm = confusion_matrix(y_true, y_pred)

print("Confusion Matrix:")
print(cm)

# Trực quan hóa confusion matrix
plt.figure(figsize=(6, 4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
            xticklabels=['Predicted Negative (0)', 'Predicted Positive (1)'], 
            yticklabels=['Actual Negative (0)', 'Actual Positive (1)'])
plt.ylabel('Actual Label')
plt.xlabel('Predicted Label')
plt.title('Confusion Matrix')
plt.show()
```

Dựa vào ma trận trên, chúng ta có:
- **TN = 3:** 3 lần mô hình dự đoán đúng là 0.
- **FP = 1:** 1 lần mô hình dự đoán là 1 nhưng thực tế là 0.
- **FN = 1:** 1 lần mô hình dự đoán là 0 nhưng thực tế là 1.
- **TP = 4:** 4 lần mô hình dự đoán đúng là 1.

### Các Chỉ số Đánh giá Cơ bản

Từ confusion matrix, chúng ta có thể tính toán các chỉ số (metrics) để đánh giá hiệu quả mô hình một cách định lượng.

#### 1. Accuracy (Độ chính xác)

**Accuracy** là chỉ số đơn giản và phổ biến nhất, đo lường tỷ lệ dự đoán đúng trên tổng số dự đoán.

$$ \text{Accuracy} = \frac{\text{TP} + \text{TN}}{\text{TP} + \text{TN} + \text{FP} + \text{FN}} = \frac{\text{Số dự đoán đúng}}{\text{Tổng số dự đoán}} $$

**Khi nào nên dùng?**
Accuracy hoạt động tốt khi các lớp trong dữ liệu của bạn **cân bằng** (số lượng mẫu của mỗi lớp gần bằng nhau).

**Vấn đề của Accuracy:**
Khi dữ liệu bị **mất cân bằng (imbalanced)**, accuracy có thể gây hiểu lầm. Ví dụ, trong một tập dữ liệu có 99% là lớp 0 và 1% là lớp 1, một mô hình "ngây thơ" luôn dự đoán lớp 0 sẽ có accuracy là 99%, nhưng nó hoàn toàn vô dụng vì không phát hiện được bất kỳ mẫu nào của lớp 1.

```python
from sklearn.metrics import accuracy_score

# Tính accuracy từ ví dụ trên
accuracy = accuracy_score(y_true, y_pred)

print(f"Accuracy: {accuracy:.2f}")

# Tính thủ công
TP = 4
TN = 3
FP = 1
FN = 1
manual_accuracy = (TP + TN) / (TP + TN + FP + FN)
print(f"Manual Accuracy: {manual_accuracy:.2f}")
```

#### 2. Precision (Độ chuẩn)

**Precision** trả lời câu hỏi: "Trong số tất cả các trường hợp được dự đoán là Positive, có bao nhiêu trường hợp thực sự là Positive?"

$$ \text{Precision} = \frac{\text{TP}}{\text{TP} + \text{FP}} $$

**Khi nào nên dùng?**
Khi chi phí của **False Positive (FP)** là cao. Tức là, bạn muốn chắc chắn rằng khi mô hình dự đoán là Positive, nó phải thực sự đúng.
- **Ví dụ:** Phân loại email spam. Nếu một email quan trọng (non-spam) bị phân loại nhầm vào hòm thư rác (FP), người dùng có thể bỏ lỡ thông tin quan trọng. Chúng ta muốn tối ưu Precision trong trường hợp này.

```python
from sklearn.metrics import precision_score

# Tính precision
precision = precision_score(y_true, y_pred)
print(f"Precision: {precision:.2f}")

# Tính thủ công
manual_precision = TP / (TP + FP)
print(f"Manual Precision: {manual_precision:.2f}")
```

#### 3. Recall (Độ phủ) / Sensitivity

**Recall** (còn gọi là **Sensitivity** hoặc **True Positive Rate**) trả lời câu hỏi: "Trong số tất cả các trường hợp thực sự là Positive, mô hình đã tìm thấy (dự đoán đúng) được bao nhiêu trường hợp?"

$$ \text{Recall} = \frac{\text{TP}}{\text{TP} + \	ext{FN}} $$

**Khi nào nên dùng?**
Khi chi phí của **False Negative (FN)** là cao. Tức là, bạn muốn tránh bỏ sót các trường hợp Positive.
- **Ví dụ:** Chẩn đoán bệnh ung thư. Bỏ sót một bệnh nhân (FN) nguy hiểm hơn nhiều so với việc chẩn đoán nhầm một người khỏe mạnh (FP). Chúng ta muốn tối ưu Recall trong trường hợp này.

```python
from sklearn.metrics import recall_score

# Tính recall
recall = recall_score(y_true, y_pred)
print(f"Recall: {recall:.2f}")

# Tính thủ công
manual_recall = TP / (TP + FN)
print(f"Manual Recall: {manual_recall:.2f}")
```

## Tóm tắt

- **Đánh giá mô hình** là bước không thể thiếu để đo lường hiệu quả và độ tin cậy của mô hình.
- **Confusion Matrix** là công cụ nền tảng để phân tích kết quả dự đoán của mô hình phân loại.
- **Accuracy** là chỉ số hữu ích khi dữ liệu cân bằng, nhưng có thể gây hiểu lầm khi dữ liệu mất cân bằng.
- **Precision** quan trọng khi chi phí của False Positive cao.
- **Recall** quan trọng khi chi phí của False Negative cao.

Trong bài học tiếp theo, chúng ta sẽ tìm hiểu sâu hơn về sự đánh đổi giữa Precision và Recall, và một chỉ số kết hợp cả hai là F1-Score.

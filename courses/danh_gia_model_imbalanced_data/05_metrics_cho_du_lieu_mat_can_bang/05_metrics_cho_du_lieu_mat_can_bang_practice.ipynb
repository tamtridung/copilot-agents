### Phần 2: Bài tập thực hành

Chúng ta sẽ làm việc với một confusion matrix cụ thể để tính toán các chỉ số.

````python
import numpy as np

# Giả sử chúng ta có một confusion matrix từ kết quả dự đoán của một mô hình
# TN = 2500, FP = 50
# FN = 100,  TP = 30
tn, fp, fn, tp = 2500, 50, 100, 30
````

**Bài tập 6:** Từ confusion matrix trên, hãy viết code để tính toán các giá trị sau:
a. Sensitivity (Recall của lớp 1)
b. Specificity (Recall của lớp 0)
c. Precision của lớp 1

````python
# Viết code của bạn ở đây
sensitivity = tp / (tp + fn)
specificity = tn / (tn + fp)
precision = tp / (tp + fp)

print(f"Sensitivity (Recall lớp 1): {sensitivity:.4f}")
print(f"Specificity (Recall lớp 0): {specificity:.4f}")
print(f"Precision lớp 1: {precision:.4f}")
````

**Bài tập 7:** Dựa vào kết quả từ Bài tập 6, hãy tính toán:
a. Balanced Accuracy
b. G-Mean

````python
# Viết code của bạn ở đây
balanced_accuracy = (sensitivity + specificity) / 2
g_mean = np.sqrt(sensitivity * specificity)

print(f"Balanced Accuracy: {balanced_accuracy:.4f}")
print(f"G-Mean: {g_mean:.4f}")
````

**Bài tập 8:** Vẫn từ các giá trị đã tính, hãy tính toán các F-beta score sau cho lớp 1:
a. F1-score (beta=1)
b. F2-score (beta=2)
c. F0.5-score (beta=0.5)

````python
# Viết code của bạn ở đây
def fbeta_score_manual(precision, recall, beta):
    if precision + recall == 0:
        return 0
    return (1 + beta**2) * (precision * recall) / ((beta**2 * precision) + recall)

f1 = fbeta_score_manual(precision, sensitivity, 1)
f2 = fbeta_score_manual(precision, sensitivity, 2)
f0_5 = fbeta_score_manual(precision, sensitivity, 0.5)

print(f"F1-Score: {f1:.4f}")
print(f"F2-Score (ưu tiên Recall): {f2:.4f}")
print(f"F0.5-Score (ưu tiên Precision): {f0_5:.4f}")
````

### Phần 3: Đáp án và Giải thích

**Câu 1:**
`Accuracy` là một chỉ số tồi vì nó bị "lấn át" bởi lớp đa số. Một mô hình "ngu ngốc" chỉ cần dự đoán tất cả các mẫu là lớp 0 (lớp đa số) sẽ đạt được `Accuracy` là 99%. Tuy nhiên, mô hình này hoàn toàn vô dụng vì nó không thể phát hiện được bất kỳ trường hợp nào của lớp 1 (lớp thiểu số), vốn là mục tiêu chính của bài toán.

**Câu 2:**
- `Balanced Accuracy` là trung bình cộng của Recall hai lớp (`(R0 + R1)/2`).
- `G-Mean` là trung bình nhân của Recall hai lớp (`sqrt(R0 * R1)`).
`G-Mean` sẽ cho giá trị thấp hơn đáng kể khi một trong hai giá trị Recall rất thấp (gần 0). Do tính chất của phép nhân, một số gần 0 sẽ kéo cả kết quả xuống gần 0. Trong khi đó, phép cộng của `Balanced Accuracy` sẽ "nhẹ tay" hơn. Ví dụ, nếu R1=0.1 và R0=0.99, `Balanced Accuracy` là `(0.1 + 0.99)/2 = 0.545`, trong khi `G-Mean` chỉ là `sqrt(0.1 * 0.99) ≈ 0.315`.

**Câu 3:**
Trong bài toán phát hiện gian lận, chúng ta ưu tiên **Recall**. Bỏ sót một giao dịch gian lận (FN) là rất nguy hiểm. Do đó, chúng ta muốn mô hình có khả năng "vợt" được càng nhiều ca gian lận càng tốt, chấp nhận việc có thể báo nhầm một vài giao dịch hợp lệ (FP).
Vì ưu tiên Recall, chúng ta sẽ chọn `beta > 1`. Một lựa chọn phổ biến và hợp lý là **`beta=2` (F2-score)**, có nghĩa là chúng ta coi trọng Recall gấp đôi Precision.

**Câu 4:**
- `Balanced Accuracy` = (0.9 + 0.9) / 2 = 0.9
- `G-Mean` = sqrt(0.9 * 0.9) = 0.9
Khi hiệu suất trên hai lớp là cân bằng, hai chỉ số này cho cùng một kết quả.

**Câu 5:**
- `Balanced Accuracy` = (0.99 + 0.1) / 2 = 0.545
- `G-Mean` = sqrt(0.99 * 0.1) ≈ 0.315
**Nhận xét:** Rõ ràng, `G-Mean` (0.315) đã "trừng phạt" mô hình nặng nề hơn nhiều so với `Balanced Accuracy` (0.545) vì hiệu suất trên lớp 1 quá tệ (Recall=0.1). Điều này cho thấy `G-Mean` rất nghiêm khắc và là một chỉ báo tốt về sự mất cân bằng trong hiệu suất của mô hình.

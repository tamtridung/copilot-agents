**Lỗi sai ở đây là gì? Hãy viết lại đoạn code theo đúng quy trình.**

````python
# Viết lại code đúng ở đây
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from imblearn.over_sampling import SMOTE
from sklearn.metrics import classification_report # Sử dụng chỉ số tốt hơn

# 1. Tạo dữ liệu
X, y = make_classification(n_samples=1000, weights=[0.9, 0.1], random_state=42)

# 2. Chia dữ liệu GỐC thành train và test TRƯỚC TIÊN
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

# 3. Áp dụng SMOTE CHỈ trên tập huấn luyện
smote = SMOTE(random_state=42)
X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)

# 4. Huấn luyện mô hình trên dữ liệu đã resample và đánh giá trên tập test GỐC
model = LogisticRegression()
model.fit(X_train_resampled, y_train_resampled)
y_pred = model.predict(X_test)

print("Kết quả đánh giá trên tập test gốc:")
print(classification_report(y_test, y_pred))
````

**Bài tập 7: Hoàn thành Pipeline với Undersampling**

Điền vào các chỗ trống (`...`) để hoàn thành quy trình sử dụng `RandomUnderSampler`.

````python
from collections import Counter
from imblearn.under_sampling import RandomUnderSampler

# Sử dụng lại X, y, X_train, X_test, y_train, y_test từ bài 6

print(f"Phân phối lớp train ban đầu: {Counter(y_train)}")

# 1. Khởi tạo RandomUnderSampler
rus = RandomUnderSampler(random_state=42)

# 2. Resample tập huấn luyện
X_train_rus, y_train_rus = rus.fit_resample(X_train, y_train)

print(f"Phân phối lớp train sau khi undersample: {Counter(y_train_rus)}")

# 3. Huấn luyện mô hình
model_rus = LogisticRegression()
model_rus.fit(X_train_rus, y_train_rus)

# 4. Đưa ra dự đoán trên tập test GỐC
y_pred_rus = model_rus.predict(X_test)

# 5. In ra báo cáo kết quả
print("\nKết quả với RandomUnderSampler:")
print(classification_report(y_test, y_pred_rus))
````

### Phần 3: Đáp án và Giải thích

**Câu 1:**
Chúng ta không áp dụng resampling trên tập test vì tập test phải đại diện cho phân phối dữ liệu **trong thực tế** mà mô hình sẽ gặp phải. Mục đích của việc đánh giá là để xem mô hình, sau khi đã được huấn luyện (có thể trên dữ liệu đã được điều chỉnh), hoạt động tốt như thế nào trong điều kiện thực tế. Nếu chúng ta thay đổi (resample) tập test, chúng ta đang đánh giá mô hình trên một kịch bản giả tạo, không phản ánh đúng hiệu suất của nó khi triển khai. Điều này được gọi là **"rò rỉ dữ liệu" (data leakage)**, vì thông tin về cấu trúc dữ liệu cân bằng đã bị "rò rỉ" vào quá trình đánh giá, làm cho kết quả trở nên lạc quan một cách sai lầm.

**Câu 2:**
Nhược điểm lớn nhất của `Random Undersampling` là **mất mát thông tin (information loss)**. Bằng cách loại bỏ ngẫu nhiên các mẫu từ lớp đa số, chúng ta có nguy cơ vứt đi những mẫu quan trọng, chứa các đặc điểm hữu ích hoặc là các ví dụ ranh giới (boundary examples) giúp mô hình xác định ranh giới quyết định một cách chính xác. Hậu quả là mô hình có thể không học được đầy đủ sự đa dạng của lớp đa số, dẫn đến hiệu suất kém hơn trên chính lớp này (giảm Recall của lớp đa số) và có thể tạo ra một ranh giới quyết định không tối ưu.

**Câu 3:**
Nhược điểm lớn nhất của `Random Oversampling` là **học vẹt (overfitting)**. Bằng cách chỉ đơn giản là sao chép các mẫu của lớp thiểu số, mô hình có thể học thuộc lòng chính những mẫu được lặp lại đó thay vì học các quy luật khái quát. Hậu quả là mô hình có thể hoạt động rất tốt trên tập huấn luyện (vì nó đã "nhớ" các mẫu) nhưng lại hoạt động kém trên tập kiểm tra hoặc dữ liệu mới mà nó chưa từng thấy, vì nó không có khả năng khái quát hóa tốt.

**Câu 4:**
SMOTE giải quyết vấn đề overfitting bằng cách không sao chép mà **tạo ra các mẫu tổng hợp (synthetic) mới**. Ý tưởng chính là: thay vì lặp lại một điểm, nó sẽ tìm các điểm lân cận của điểm đó (cũng thuộc lớp thiểu số) và tạo ra một điểm mới nằm trên đường thẳng nối giữa chúng. Điều này tạo ra các mẫu mới, hợp lý, nằm gần các mẫu hiện có, giúp mở rộng vùng quyết định của lớp thiểu số một cách mượt mà hơn và buộc mô hình phải khái quát hóa thay vì học thuộc lòng.

**Câu 5:**
Bạn có thể cân nhắc sử dụng `Random Undersampling` khi **tập dữ liệu của bạn cực kỳ lớn**. Trong trường hợp này, lớp đa số có thể có hàng triệu mẫu và rất dư thừa. Việc loại bỏ một phần lớn trong số chúng có thể không gây mất mát quá nhiều thông tin quan trọng nhưng lại giúp **giảm đáng kể thời gian và chi phí tính toán** để huấn luyện mô hình. Đây là một sự đánh đổi thực tế giữa hiệu suất và tài nguyên.

**Câu hỏi 3:**
```python
class Manager(Employee):
    def __init__(self, name, employee_id, base_salary, bonus):
        # Gọi __init__ của class cha
        super().__init__(name, employee_id, base_salary)
        # Thêm thuộc tính mới
        self.bonus = bonus

    # Ghi đè phương thức tính lương
    def calculate_monthly_salary(self):
        return self.base_salary + self.bonus
```

**Câu hỏi 5:**
```python
class Developer(Employee):
    def __init__(self, name, employee_id, base_salary, programming_language):
        super().__init__(name, employee_id, base_salary)
        self.programming_language = programming_language

    def show_language(self):
        print(f"{self.name} lập trình bằng ngôn ngữ {self.programming_language}.")

    def calculate_monthly_salary(self):
        # Gọi phương thức gốc để lấy lương cơ bản
        salary = super().calculate_monthly_salary()
        if self.programming_language.lower() == 'python':
            print(f"(Thưởng 10% cho lập trình viên Python!)")
            salary *= 1.1 # Thưởng 10%
        return salary
```

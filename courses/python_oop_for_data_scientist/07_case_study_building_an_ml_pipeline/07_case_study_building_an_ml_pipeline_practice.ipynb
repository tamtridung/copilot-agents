**Answer to Question 1:**

An Abstract Base Class (ABC) is a class that cannot be instantiated on its own. Its purpose is to define a common interface (a set of methods and properties) that all of its subclasses must implement.

In our pipeline, the `PipelineStep` ABC served as a formal contract. It guaranteed that every component (`DataLoader`, `StandardScalerStep`, etc.) would have a `.process()` method. This makes the main `Pipeline` class simpler and more reliable, as it knows it can call `.process()` on any step object it is given, demonstrating polymorphism.

---

**Answer to Question 2:**

```python
from abc import ABC, abstractmethod

class TextProcessorStep(ABC):
    @abstractmethod
    def run(self, text):
        """All text processing steps must implement this."""
        pass
```

---

**Answer to Question 3:**

```python
class TextLoader(TextProcessorStep):
    def run(self, text=None):
        print("Step 1: Loading text...")
        return "This is the sample TEXT for our Pipeline."

# --- Verification ---
loader = TextLoader()
print(loader.run())
```

**Answer to Question 5:**

```python
class WordCounter(TextProcessorStep):
    def run(self, text):
        print("Step 3: Counting words...")
        return len(text.split())

class CharCounter(TextProcessorStep):
    def run(self, text):
        print("Step 3: Counting characters...")
        return len(text)

# --- Verification ---
sample_text = "hello world"
word_analyzer = WordCounter()
char_analyzer = CharCounter()
print(f"Word count: {word_analyzer.run(sample_text)}")
print(f"Char count: {char_analyzer.run(sample_text)}")
```

---

**Answer to Question 6:**

```python
class TextPipeline:
    def __init__(self, steps):
        if not all(isinstance(s, TextProcessorStep) for s in steps):
            raise TypeError("All steps must be a TextProcessorStep")
        self.steps = steps
        
    def process(self):
        data = None
        for step in self.steps:
            data = step.run(data)
        return data
```

---

**Answer to Question 7:**

```python
# --- Pipeline 1: Word Counting ---
print("--- Running Word Count Pipeline ---")
word_pipeline_steps = [
    TextLoader(),
    LowercaseTransformer(),
    WordCounter() # Use the word counter
]
word_pipeline = TextPipeline(steps=word_pipeline_steps)
result1 = word_pipeline.process()
print(f"Final Result (Word Count): {result1}")


# --- Pipeline 2: Character Counting ---
print("\n--- Running Character Count Pipeline ---")
char_pipeline_steps = [
    TextLoader(),
    LowercaseTransformer(),
    CharCounter() # Swap in the character counter
]
char_pipeline = TextPipeline(steps=char_pipeline_steps)
result2 = char_pipeline.process()
print(f"Final Result (Character Count): {result2}")
```

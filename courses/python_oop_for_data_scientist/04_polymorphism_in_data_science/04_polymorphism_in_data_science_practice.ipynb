**Answer to Question 1:**

**Duck typing** is Python's approach to polymorphism. It's a concept where the type or class of an object is less important than the methods it defines. If an object has the methods and attributes required for a certain operation (i.e., it 'walks and quacks like a duck'), it can be used in that operation, regardless of its specific class.

**Example:** Imagine you have a job opening for a writer. You don't care if the applicant has a degree in English, Journalism, or Creative Writing (their 'class'). You just care if they can perform the `write_article()` method. As long as they can do the job, you can hire them. Your 'hiring' function is polymorphic; it works with any person who has the `write_article` skill.

**Answer to Question 3:**

```python
class EmailNotifier:
    def send(self, message):
        print(f"Sending EMAIL: '{message}'")

class SmsNotifier:
    def send(self, message):
        print(f"Sending SMS: '{message}'")

class PushNotifier:
    def send(self, message):
        print(f"Sending PUSH NOTIFICATION: '{message}'")

def send_alert(message, notifiers):
    print("--- Broadcasting Alert ---")
    for notifier in notifiers:
        notifier.send(message)

# --- Verification ---
all_notifiers = [EmailNotifier(), SmsNotifier(), PushNotifier()]
alert_message = "System is going down for maintenance!"

send_alert(alert_message, all_notifiers)
```

**Answer to Question 5:**

```python
# 1. Parent class
class Document:
    def __init__(self, filename):
        self.filename = filename
    
    # 2. Abstract search method
    def search(self, keyword):
        raise NotImplementedError("Subclasses must implement the search method.")

# 3. Child classes
class PdfDocument(Document):
    def search(self, keyword):
        print(f"Searching for '{keyword}' in PDF document: {self.filename}")

class WordDocument(Document):
    def search(self, keyword):
        print(f"Searching for '{keyword}' in Word document: {self.filename}")

# 4. Create a list of documents
documents = [
    PdfDocument("report.pdf"),
    WordDocument("notes.docx")
]

# 5. Write a polymorphic loop
for doc in documents:
    doc.search('python')
```

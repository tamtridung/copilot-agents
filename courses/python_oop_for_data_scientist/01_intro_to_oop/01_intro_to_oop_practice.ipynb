**Answer to Question 1:**

A **class** is a blueprint or template. It defines the structure and behavior (attributes and methods) that all objects of that type will have, but it isn't a usable entity on its own. For example, an architect's blueprint for a house.

An **object** is a specific, concrete instance created from that class. It's the actual thing you can interact with. For example, a physical house built using the architect's blueprint. You can build many different houses (objects) from the same blueprint (class).

**Answer to Question 7:**

**The error:** The `start_engine` method is missing the `self` parameter.

**Why it happens:** In Python, when you call a method on an object (e.g., `my_car.start_engine()`), Python automatically passes the object itself (`my_car`) as the first argument to the method. The method definition must have a parameter to receive this object. By convention, this parameter is always named `self`. Without it, Python thinks you're calling a function that takes zero arguments but you're trying to pass one (the object itself), leading to a `TypeError`.

**How to fix it:** Add `self` as the first parameter of the `start_engine` method.

```python
# Corrected Code
class Car:
    def __init__(self, make, model):
        self.make = make
        self.model = model

    # Add 'self' here
    def start_engine(self):
        print(f"The {self.make} {self.model}'s engine is running.")

my_car = Car("Toyota", "Camry")
my_car.start_engine()
```

**Answer to Question 8:**

````python
# 1. Create the class
class DataFile:
    def __init__(self, filename, file_type):
        self.filename = filename
        self.file_type = file_type
        self.size_mb = None # 2. Initialize size_mb to None

    # 3. Create the update_size method
    def update_size(self, new_size):
        self.size_mb = new_size
        print(f"Size of '{self.filename}' updated to {self.size_mb} MB.")

# 4. Create an object
patient_file = DataFile(filename="patient_records.csv", file_type="csv")

# 5. Call the update_size method
patient_file.update_size(15.5)

# 6. Print the attributes to confirm
print("\n--- File Details ---")
print(f"Filename: {patient_file.filename}")
print(f"File Type: {patient_file.file_type}")
print(f"Size (MB): {patient_file.size_mb}")
````

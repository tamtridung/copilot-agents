# Đồ án cuối khóa: Xây dựng Framework Chạy Thử nghiệm Machine Learning

## Bối cảnh

Trong vai trò một Data Scientist, một trong những công việc thường xuyên nhất của bạn là thử nghiệm nhiều mô hình machine learning khác nhau trên một bộ dữ liệu để tìm ra mô hình hoạt động tốt nhất. Quá trình này thường bao gồm các bước lặp đi lặp lại: chuẩn bị dữ liệu, huấn luyện mô hình, đánh giá mô hình, và ghi lại kết quả.

Việc viết code cho mỗi lần thử nghiệm từ đầu rất tốn thời gian, dễ gây lỗi và khó so sánh kết quả một cách công bằng. Nếu bạn có thể xây dựng một "framework" nhỏ, tái sử dụng được, cho phép bạn chỉ cần định nghĩa một lần và chạy cho nhiều mô hình, hiệu suất công việc của bạn sẽ tăng lên đáng kể.

## Mục tiêu đồ án

Nhiệm vụ của bạn là **sử dụng tất cả các kiến thức về Lập trình Hướng đối tượng (OOP) đã học để xây dựng một framework nhỏ, linh hoạt cho việc chạy và so sánh các thử nghiệm machine learning.**

Framework của bạn phải có khả năng:
1.  Đọc dữ liệu từ một nguồn cho trước.
2.  Áp dụng một chuỗi các bước tiền xử lý có thể cấu hình được.
3.  Huấn luyện nhiều mô hình khác nhau trên cùng một dữ liệu đã xử lý.
4.  Đánh giá mỗi mô hình bằng một bộ các số liệu (metrics) được định nghĩa trước.
5.  In ra một bảng tóm tắt kết quả để dễ dàng so sánh hiệu suất của các mô hình.

## Yêu cầu về thiết kế (OOP)

Đây là phần quan trọng nhất của đồ án. Bạn phải áp dụng các khái niệm OOP sau:

-   **Classes:** Mọi thành phần chính của framework phải là một class (ví dụ: `ExperimentRunner`, `DataLoader`, `Preprocessor`, `ModelWrapper`, `Evaluator`).
-   **Inheritance & Polymorphism:**
    -   Tạo một class cha trừu tượng `BaseProcessorStep` cho các bước tiền xử lý.
    -   Tạo một class cha trừu tượng `BaseModel` cho tất cả các mô hình. Các mô hình cụ thể (ví dụ `SklearnModel`) sẽ kế thừa từ nó.
-   **Encapsulation:** Che giấu logic nội bộ. Ví dụ, `ExperimentRunner` không cần biết *cách* một mô hình được huấn luyện, nó chỉ cần gọi phương thức `train()` của mô hình đó.
-   **Design Patterns:**
    -   **Factory Pattern:** Tạo một `ModelFactory` để có thể tạo ra các instance của mô hình (`SklearnModel`) dựa trên một chuỗi định danh (ví dụ: `'logistic_regression'`).
    -   **Strategy Pattern:** Các bước tiền xử lý (ví dụ: `Imputer`, `Scaler`) có thể được xem như các "chiến lược" và được đưa vào một pipeline xử lý chính.
    -   **(Tùy chọn nâng cao) Singleton Pattern:** Tạo một class `Config` để quản lý toàn bộ cấu hình của thử nghiệm (ví dụ: đường dẫn file, danh sách mô hình cần chạy, các tham số,...) và đảm bảo chỉ có một đối tượng config duy nhất trong toàn bộ chương trình.

## Các bước gợi ý để hoàn thành

1.  **Thiết kế trên giấy:** Trước khi code, hãy vẽ ra sơ đồ các class, mối quan hệ kế thừa và cách chúng tương tác với nhau. Việc này sẽ giúp bạn tiết kiệm rất nhiều thời gian.

2.  **Xây dựng các Class lõi:**
    -   `BaseModel` (ABC) với các phương thức trừu tượng `train(X, y)` và `predict(X)`.
    -   `SklearnModel(BaseModel)`: Một class cụ thể để bọc (wrap) các mô hình từ scikit-learn. `__init__` của nó nên nhận một instance của mô hình sklearn.
    -   `ModelFactory`: Class có phương thức `create_model(model_name, **params)` để trả về một instance của `SklearnModel`.

3.  **Xây dựng các Class xử lý dữ liệu:**
    -   `DataLoader`: Class có phương thức `load(path, column_names)` để tải dữ liệu.
    -   `DataPreprocessor`: Class có phương thức `process(data)` để thực hiện các bước như chia train/test, xử lý giá trị thiếu (nhiều cột có giá trị 0 không hợp lệ, ví dụ `Glucose`, `BloodPressure`), và scaling dữ liệu.

4.  **Xây dựng Class đánh giá:**
    -   `Evaluator`: Class có phương thức `evaluate(y_true, y_pred)` để tính toán một loạt các metrics (ví dụ: `accuracy`, `precision`, `recall`, `f1-score`) và trả về một dictionary kết quả.

5.  **Xây dựng Class "Nhạc trưởng" `ExperimentRunner`:**
    -   Đây là class quan trọng nhất, kết nối mọi thứ lại với nhau.
    -   `__init__` của nó có thể nhận vào một cấu hình (ví dụ: một dictionary) định nghĩa toàn bộ thử nghiệm.
    -   Phương thức `run()` của nó sẽ thực hiện toàn bộ quy trình:
        a. Gọi `DataLoader`.
        b. Gọi `DataPreprocessor`.
        c. Lặp qua danh sách các mô hình cần chạy trong cấu hình:
           i. Dùng `ModelFactory` để tạo mô hình.
           ii. Gọi `model.train()`.
           iii. Gọi `model.predict()`.
           iv. Gọi `Evaluator`.
           v. Lưu kết quả của mô hình này.
        d. In ra bảng so sánh kết quả cuối cùng.

6.  **Tạo file chạy chính:**
    -   Tạo một dictionary `config` để định nghĩa thử nghiệm của bạn. Ví dụ:
      ```python
      config = {
          'data_url': '...', 
          'column_names': [...],
          'models_to_run': {
              'logistic_regression': {'C': 1.0},
              'decision_tree': {'max_depth': 5},
              'random_forest': {'n_estimators': 100, 'max_depth': 5}
          },
          'evaluation_metrics': ['accuracy', 'precision', 'recall', 'f1']
      }
      ```
    -   Tạo một instance của `ExperimentRunner` với `config` này và gọi `runner.run()`.

### Question 6: Arithmetic Magic Method `__add__`

1. Check if the `other` vector has the same length as `self`. If not, raise a `ValueError`.

```python
class Vector:
    def __init__(self, values):
        self.values = values
        
    def __repr__(self):
        return f"Vector({self.values})"
    
    def __len__(self):
        return len(self.values)
    
    # Add the __add__ method here

# --- Verification ---
v1 = Vector([1, 2, 3])
v2 = Vector([10, 20, 30])

# This should call your __add__ method
v3 = v1 + v2 
print(v3) # Should print Vector([11, 22, 33])

# This should raise a ValueError
v4 = Vector([1, 2])
try:
    v1 + v4
except ValueError as e:
    print(f"\nSuccess! Error caught: {e}")
```

**Answer to Question 1:**

- `__str__`: Its goal is to be **readable**. It's called by the `print()` function and `str()` constructor. It should return a user-friendly string that gives a nice summary of the object's state.
- `__repr__`: Its goal is to be **unambiguous and informative**. It's called when you inspect an object in a terminal or notebook (by just typing its name). It should return a developer-friendly string that, ideally, could be used to recreate the object. It's the fallback for `__str__` if `__str__` is not defined.

**Answer to Question 6:**

```python
class Vector:
    def __init__(self, values):
        self.values = values
        
    def __repr__(self):
        return f"Vector({self.values})"
    
    def __len__(self):
        return len(self.values)
    
    def __add__(self, other):
        # 1. Check length
        if len(self) != len(other):
            raise ValueError("Vectors must be the same length to be added.")
        
        # 2. Create new list with sums
        # Using a zip for a clean, Pythonic loop
        new_values = [s + o for s, o in zip(self.values, other.values)]
        
        # 3. Return a new Vector object
        return Vector(new_values)

# --- Verification ---
v1 = Vector([1, 2, 3])
v2 = Vector([10, 20, 30])

v3 = v1 + v2 
print(v3)

v4 = Vector([1, 2])
try:
    v1 + v4
except ValueError as e:
    print(f"\nSuccess! Error caught: {e}")
```

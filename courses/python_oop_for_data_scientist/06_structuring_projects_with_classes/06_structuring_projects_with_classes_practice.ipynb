**Answer to Question 1:**

The Single Responsibility Principle (SRP) states that a class should have only one job or one reason to change. 

In a data science project, this is useful because it helps break down a complex workflow (e.g., data loading, cleaning, training, visualization) into manageable, independent parts. This makes the code more modular, easier to test (you can test just the cleaning part), easier to reuse (you can reuse the data loader for another project), and easier to modify (you can swap out a model without touching the data cleaning code).

**Answer to Question 4:**

```python
class FileSaver:
    def save(self, content, filename):
        """Saves the given content to a file."""
        with open(filename, 'w') as f:
            f.write(content)
        print(f"Content saved to {filename}")

class Report:
    def __init__(self, title, content, saver):
        self.title = title
        self.content = content
        self.saver = saver # Store the saver object
        
    def format(self):
        return f"--- {self.title} ---\n{self.content}"
    
    def save(self, filename):
        """Delegates the saving task to the saver object."""
        formatted_content = self.format()
        self.saver.save(formatted_content, filename)

# --- Verification ---
my_saver = FileSaver()
my_report = Report(
    title="Q3 Sales", 
    content="Sales are up by 15%.", 
    saver=my_saver
)

my_report.save("sales_report.txt")
```

**Answer to Question 5:**

```python
import string

class TextLoader:
    def __init__(self, text_content):
        self._text = text_content
    
    def load(self):
        return self._text

class TextCleaner:
    def process(self, text):
        lower_text = text.lower()
        # Create a translation table to remove punctuation
        translator = str.maketrans('', '', string.punctuation)
        return lower_text.translate(translator)

class TextAnalyzer:
    def analyze(self, text):
        words = text.split()
        return len(words)

class TextProcessor:
    def __init__(self, loader, cleaner, analyzer):
        self.loader = loader
        self.cleaner = cleaner
        self.analyzer = analyzer
        
    def run(self):
        print("Starting text processing pipeline...")
        # 1. Load
        raw_text = self.loader.load()
        print(f"Loaded text: '{raw_text[:30]}...' ")
        
        # 2. Clean
        cleaned_text = self.cleaner.process(raw_text)
        print(f"Cleaned text: '{cleaned_text[:30]}...' ")
        
        # 3. Analyze
        result = self.analyzer.analyze(cleaned_text)
        print("Analysis complete.")
        return result

# --- Verification ---
raw_text = "Hello World! This is an Example of OOP design. It is powerful."

loader = TextLoader(raw_text)
cleaner = TextCleaner()
analyzer = TextAnalyzer()

processor = TextProcessor(loader, cleaner, analyzer)
word_count = processor.run()

print(f"\nThe final word count is: {word_count}")
```

## Tại sao OOP lại quan trọng với Data Scientist?

Khi mới bắt đầu, chúng ta thường viết code Python dưới dạng các script đơn giản, chạy từ trên xuống dưới. Điều này rất hiệu quả cho các tác vụ nhỏ, riêng lẻ. Tuy nhiên, khi các dự án data science lớn dần lên, bạn sẽ phải đối mặt với những thách thức:

- **Code lặp lại:** Bạn có thể phải copy-paste các đoạn code để xử lý dữ liệu, huấn luyện mô hình ở nhiều nơi.
- **Khó quản lý:** Một file script dài hàng trăm dòng sẽ rất khó để đọc, sửa lỗi và nâng cấp.
- **Khó tái sử dụng:** Việc mang một phần logic từ dự án này sang dự án khác rất phức tạp.

Lập trình hướng đối tượng (Object-Oriented Programming - OOP) giải quyết những vấn đề này bằng cách giúp chúng ta **tổ chức code** theo một cách logic hơn, xoay quanh các "đối tượng" (objects) thay vì các hàm và thủ tục rời rạc. Nó cho phép chúng ta:

- **Đóng gói (Encapsulation):** Gom dữ liệu và các hàm xử lý dữ liệu đó vào chung một nơi gọi là "class".
- **Tái sử dụng (Reusability):** Dễ dàng sử dụng lại các class đã viết cho các dự án khác nhau.
- **Dễ bảo trì (Maintainability):** Code được chia thành các module nhỏ, độc lập, giúp việc sửa lỗi và thêm tính năng mới trở nên đơn giản hơn.

## Giải thích trực quan: Khuôn bánh và Những cái bánh

Để hiểu về class và object, hãy tưởng tượng bạn có một **khuôn làm bánh** (ví dụ: khuôn hình ngôi sao).

- **Class (Lớp):** Cái **khuôn bánh** chính là `class`. Nó định nghĩa hình dạng, kích thước, và các đặc điểm chung mà tất cả các cái bánh làm từ khuôn này sẽ có. Bản thân cái khuôn không phải là một cái bánh, nó chỉ là bản thiết kế.
- **Object (Đối tượng):** Mỗi **cái bánh** bạn tạo ra từ khuôn đó là một `object` (hay còn gọi là `instance`). Mỗi cái bánh là một thực thể riêng biệt, nhưng chúng đều tuân theo thiết kế của cái khuôn. Bạn có thể tạo ra nhiều cái bánh (nhiều object) từ cùng một khuôn (cùng một class).

## Các khái niệm cốt lõi

Bây giờ, hãy đi vào các khái niệm cụ thể trong Python.

### 1. Class

Một `class` là một bản thiết kế để tạo ra các object. Chúng ta dùng từ khóa `class` để định nghĩa một class mới.

````python
# Cú pháp định nghĩa một class đơn giản
# Tên class thường được viết theo quy tắc PascalCase (viết hoa chữ cái đầu mỗi từ)
class Dog:
    pass # pass có nghĩa là chưa làm gì cả
````

Class `Dog` ở trên rất đơn giản và chưa có gì bên trong. Nó giống như một bản thiết kế trống.

### 2. Object (Instance)

Để tạo một object (instance) từ một class, chúng ta gọi tên class đó như gọi một hàm.

````python
# Tạo ra hai object (hai chú chó) từ class Dog
my_dog = Dog()
your_dog = Dog()

print(my_dog)
print(your_dog)
````

Bạn có thể thấy `my_dog` và `your_dog` là hai object khác nhau, được lưu ở hai địa chỉ bộ nhớ khác nhau. Chúng là hai "cái bánh" riêng biệt được tạo ra từ cùng một "khuôn" `Dog`.

### 3. Constructor (`__init__`)

Khi tạo một object, chúng ta thường muốn nó có sẵn một vài thuộc tính ban đầu (ví dụ: một chú chó khi sinh ra đã có tên và tuổi). Phương thức `__init__` (gọi là constructor) là một phương thức đặc biệt, nó sẽ **tự động được gọi** mỗi khi một object mới được tạo ra.

Tham số đầu tiên của `__init__` (và của tất cả các phương thức trong class) luôn là `self`. `self` đại diện cho chính object đang được tạo ra hoặc đang gọi phương thức đó.

````python
class Dog:
    # Constructor: được gọi khi tạo object mới
    def __init__(self, name, age):
        # Gán các giá trị đầu vào cho các thuộc tính (attributes) của object
        # self.name là thuộc tính 'name' của object
        # name là tham số được truyền vào
        self.name = name
        self.age = age
        print(f"Một chú chó mới tên là {self.name} đã được tạo!")

# Bây giờ khi tạo object, chúng ta cần truyền vào các tham số cho __init__
dog1 = Dog("Milo", 2)
dog2 = Dog("Buddy", 4)
````

### 4. Attributes (Thuộc tính)

Attributes là các biến được gắn với một object, lưu trữ dữ liệu của object đó. Trong ví dụ trên, `name` và `age` là các attributes. Chúng ta truy cập chúng bằng dấu chấm `.`.

````python
# Truy cập thuộc tính của mỗi object
print(f"{dog1.name} năm nay {dog1.age} tuổi.")
print(f"{dog2.name} năm nay {dog2.age} tuổi.")
````

### 5. Methods (Phương thức)

Methods là các hàm được định nghĩa bên trong một class. Chúng định nghĩa hành vi, tức là những gì object có thể làm. Giống như `__init__`, tham số đầu tiên của một method luôn là `self`.

````python
class Dog:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    # Đây là một method
    def bark(self):
        print(f"{self.name} đang sủa: Gâu gâu!")

    # Một method khác
    def celebrate_birthday(self):
        self.age += 1
        print(f"Chúc mừng sinh nhật {self.name}! Bây giờ bạn đã {self.age} tuổi.")

# Tạo object
my_pet = Dog("Lucy", 3)

# Gọi method của object
my_pet.bark()

print(f"Tuổi ban đầu: {my_pet.age}")
my_pet.celebrate_birthday()
print(f"Tuổi sau sinh nhật: {my_pet.age}")
````

## Ví dụ thực tế trong Khoa học dữ liệu

Giả sử chúng ta thường xuyên làm việc với các bộ dữ liệu. Thay vì truyền các DataFrame qua lại giữa các hàm, chúng ta có thể tạo một class `Dataset` để đóng gói DataFrame và các thông tin liên quan.

````python
import pandas as pd
import numpy as np

class SimpleDataset:
    """Một class đơn giản để đóng gói một DataFrame và các thông tin cơ bản."""
    
    def __init__(self, data, name, description):
        # Kiểm tra đầu vào là một DataFrame
        if not isinstance(data, pd.DataFrame):
            raise TypeError("Đầu vào 'data' phải là một pandas DataFrame.")
            
        self.data = data
        self.name = name
        self.description = description
        print(f"Dataset '{self.name}' đã được tạo.")

    def get_summary(self):
        """Trả về bảng tóm tắt thống kê của dữ liệu."""
        print(f"Tóm tắt thống kê cho {self.name}:")
        return self.data.describe()

    def get_shape(self):
        """Trả về kích thước (số hàng, số cột) của dữ liệu."""
        return self.data.shape
    
    def show_head(self, n=5):
        """Hiển thị n dòng đầu tiên."""
        return self.data.head(n)
````

Bây giờ, hãy sử dụng class `SimpleDataset` này.

````python
# Tạo một DataFrame mẫu
sample_df = pd.DataFrame({
    'age': np.random.randint(20, 60, 100),
    'salary': np.random.randint(500, 2000, 100) * 10,
    'department': np.random.choice(['HR', 'Engineering', 'Sales', 'Marketing'], 100)
})

# Tạo một object từ class SimpleDataset
employee_data = SimpleDataset(
    data=sample_df, 
    name="Employee Stats", 
    description="Dữ liệu nhân khẩu học và lương của nhân viên."
)
````

````python
# Sử dụng các method của object
print(f"Kích thước của dataset: {employee_data.get_shape()}")
````

````python
# Xem 5 dòng đầu
employee_data.show_head()
````

````python
# Lấy bảng tóm tắt thống kê
employee_data.get_summary()
````

Với class `SimpleDataset`, code của chúng ta trở nên gọn gàng và có ý nghĩa hơn. Thay vì gọi `sample_df.describe()`, chúng ta gọi `employee_data.get_summary()`. Mọi thứ liên quan đến bộ dữ liệu này (dữ liệu, tên, mô tả, các hành động) đều được gói gọn trong object `employee_data`.

## Lỗi thường gặp và cách tránh

1. **Quên `self` khi định nghĩa method**

Đây là lỗi phổ biến nhất khi mới bắt đầu. Mọi method trong class đều phải có `self` là tham số đầu tiên.

```python
# LỖI
class MyClass:
    def do_something(): # Thiếu self
        print("Doing something")

# Khi bạn gọi `obj = MyClass()` và `obj.do_something()` sẽ báo lỗi TypeError.
```

2. **Gọi method trên Class thay vì trên Object**

Methods được thiết kế để hoạt động trên dữ liệu của một object cụ thể. Bạn phải tạo object trước rồi mới gọi method.

```python
# LỖI
Dog.bark() # Sai! Phải là object của Dog để sủa, chứ không phải class Dog

# ĐÚNG
my_dog = Dog("Milo", 2)
my_dog.bark()
```

## Tóm tắt

- **Class** là bản thiết kế, là cái khuôn.
- **Object** là một thực thể cụ thể được tạo ra từ class, là cái bánh.
- Dùng `class TenClass:` để định nghĩa một class.
- `__init__` là constructor, được gọi tự động khi tạo object mới để thiết lập các thuộc tính ban đầu.
- **Attributes** là dữ liệu của object (ví dụ: `dog.name`).
- **Methods** là hành vi của object (ví dụ: `dog.bark()`).
- `self` là tham số bắt buộc đầu tiên của mọi method, đại diện cho chính object đó.
- OOP giúp tổ chức code tốt hơn, đặc biệt hữu ích cho các dự án data science phức tạp.

## Bài học tiếp theo

Trong bài học tiếp theo, chúng ta sẽ tìm hiểu về một trong những khái niệm mạnh mẽ nhất của OOP: **Kế thừa (Inheritance)**. Kế thừa cho phép một class có thể "thừa hưởng" lại các thuộc tính và phương thức từ một class khác, giúp chúng ta tái sử dụng code ở một cấp độ cao hơn.

**Câu hỏi 8:**

**Giải thích lỗi:**
1.  **Thiếu `self`:** Cả `__init__` và `get_info` đều thiếu tham số `self` ở vị trí đầu tiên. Trong Python, mọi phương thức của một instance phải có `self` để tham chiếu đến chính object đó.
2.  **Sử dụng `self` không đúng cách:** Bên trong `__init__`, code đang cố gắng gán `self.name = name` trong khi `self` chưa được định nghĩa là một tham số.

**Code đã sửa:**
```python
class Student:
    # Sửa 1: Thêm 'self' làm tham số đầu tiên
    def __init__(self, name, student_id):
        self.name = name
        self.student_id = student_id

    # Sửa 2: Thêm 'self' làm tham số đầu tiên
    def get_info(self):
        return f"Student Name: {self.name}, ID: {self.student_id}"

student1 = Student("Alice", "S123")
info = student1.get_info()
print(info)
```

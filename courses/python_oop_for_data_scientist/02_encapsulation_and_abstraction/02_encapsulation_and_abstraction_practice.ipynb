**Answer to Question 1:**

**Encapsulation** is the bundling of data (attributes) and the methods that operate on that data into a single unit (a class). It's also about *protecting* that data by controlling access to it, preventing direct, uncontrolled modification from outside the class. Think of it as building a protective wall around your data, with methods acting as secure gates.

**Abstraction** is the practice of hiding complex implementation details and exposing only a simple, relevant interface to the user. The user interacts with a high-level, easy-to-use set of methods and properties, without needing to know the messy details of how they work internally. Think of a car's dashboard—it abstracts away the engine's complexity.

They are both important because together they lead to code that is more **robust** (encapsulation prevents bugs from invalid data states) and **maintainable** (abstraction allows you to change the internal implementation without breaking the code of people who use your class).

````python
class Report:
    def __init__(self, score_a, score_b):
        self._score_a = score_a
        self._score_b = score_b

    @property
    def final_score(self):
        """Calculates and returns the average of the two scores."""
        return (self._score_a + self._score_b) / 2

# --- Verification ---
my_report = Report(80, 90)
print(f"Final score: {my_report.final_score}")

try:
    my_report.final_score = 100
except AttributeError as e:
    print(f"Success! Error caught: {e}")
````

**Answer to Question 4:**

````python
class DataColumn:
    ALLOWED_TYPES = ['numeric', 'categorical', 'text']

    def __init__(self, name, initial_type):
        self.name = name
        self.__dtype = None
        self.data_type = initial_type

    @property
    def data_type(self):
        return self.__dtype

    @data_type.setter
    def data_type(self, new_type):
        if new_type not in self.ALLOWED_TYPES:
            print(f"Error: Invalid data type '{new_type}'. Must be one of {self.ALLOWED_TYPES}")
        else:
            self.__dtype = new_type
            print(f"Data type for '{self.name}' set to '{self.__dtype}'.")

# --- Verification ---
col = DataColumn('age', 'numeric')
print(f"\nInitial type from getter: {col.data_type}")

print("\n--- Setting a valid type ---")
col.data_type = 'categorical'
print(f"New type from getter: {col.data_type}")

print("\n--- Setting an invalid type ---")
col.data_type = 'boolean'
print(f"Type after invalid attempt: {col.data_type}")
````

**Answer to Question 5:**

There is no actual error in the code—it runs because **name mangling does not prevent access from within the class itself**. The `add_entry` and `view_log` methods are part of the `Logger` class, so they can access `self.__log` without any issue. Python mangles the name to `_Logger__log`, and it uses this mangled name consistently for all references *inside the class definition*.

The key misunderstanding is thinking that `__log` is somehow private *from other methods of the same class*. It is not. Name mangling's primary purpose is to prevent accidental name clashes in subclasses (a topic for the next lesson), not to create truly private variables. The code is perfectly fine and demonstrates the correct way to use a name-mangled attribute within its own class.

**Answer to Question 6:**

````python
class DatabaseConnector_After:
    def __init__(self, connection_string):
        self._connection_string = connection_string
        self._is_connected = False

    def _connect(self):
        print(f"Internal: Connecting to {self._connection_string}...")
        self._is_connected = True

    def _fetch_data(self, query):
        if not self._is_connected:
            print("Error: Not connected to database.")
            return None
        print(f"Internal: Executing query: {query}")
        return {"user_id": 123, "name": "Alice"} # Dummy data

    def _close(self):
        print("Internal: Closing connection.")
        self._is_connected = False

    # This is the abstracted public method
    def get_user_data(self, user_id):
        """Connects, fetches user data, and closes the connection."""
        self._connect()
        query = f"SELECT * FROM users WHERE id = {user_id}"
        data = self._fetch_data(query)
        self._close()
        return data

# --- Verification ---
print("--- Using the abstracted interface ---")
db = DatabaseConnector_After("my_db_connection_string")
user_info = db.get_user_data(123)
print(f"\nReceived data: {user_info}")
````

## Đáp án

**Câu hỏi 1:**
```python
from abc import ABC, abstractmethod

class ClassificationStep(ABC):
    @abstractmethod
    def run(self, *args, **kwargs):
        pass
```

**Câu hỏi 2:**
```python
from sklearn.datasets import load_iris

class IrisLoader(ClassificationStep):
    def run(self):
        print("1. Loading Iris dataset...")
        iris = load_iris()
        X, y = iris.data, iris.target
        return X, y
```

**Câu hỏi 3:**
```python
from sklearn.model_selection import train_test_split

class DataSplitter(ClassificationStep):
    def __init__(self, test_size=0.3, random_state=42):
        self.test_size = test_size
        self.random_state = random_state

    def run(self, X, y):
        print(f"2. Splitting data (test_size={self.test_size})...")
        return train_test_split(X, y, test_size=self.test_size, random_state=self.random_state)
```

**Câu hỏi 4:**
```python
class ClassifierTrainer(ClassificationStep):
    def __init__(self, model):
        self.model = model

    def run(self, X_train, y_train):
        print(f"3. Training {self.model.__class__.__name__} model...")
        self.model.fit(X_train, y_train)
        return self.model
```

**Câu hỏi 5:**
```python
from sklearn.metrics import accuracy_score

class ModelEvaluator(ClassificationStep):
    def __init__(self, model):
        self.model = model

    def run(self, X_test, y_test):
        print(f"4. Evaluating {self.model.__class__.__name__} model...")
        predictions = self.model.predict(X_test)
        accuracy = accuracy_score(y_test, predictions)
        print(f"==> Accuracy: {accuracy:.4f}")
        return accuracy
```

**Câu hỏi 6:**
```python
from sklearn.linear_model import LogisticRegression

print("--- Running Pipeline with Logistic Regression ---")
# Step 1: Load data
loader = IrisLoader()
X, y = loader.run()

# Step 2: Split data
splitter = DataSplitter(test_size=0.3, random_state=42)
X_train, X_test, y_train, y_test = splitter.run(X, y)

# Step 3: Train model
lr_model = LogisticRegression(max_iter=200)
trainer = ClassifierTrainer(model=lr_model)
trained_lr_model = trainer.run(X_train, y_train)

# Step 4: Evaluate model
evaluator = ModelEvaluator(model=trained_lr_model)
accuracy = evaluator.run(X_test, y_test)
```

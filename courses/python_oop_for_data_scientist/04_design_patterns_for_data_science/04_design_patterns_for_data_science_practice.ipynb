```python
import pandas as pd
from abc import ABC, abstractmethod

class DataWriter(ABC):
    @abstractmethod
    def write(self, data: pd.DataFrame, path: str):
        pass

class CsvWriter(DataWriter):
    def write(self, data: pd.DataFrame, path: str):
        print(f"Đang ghi {len(data)} dòng ra file CSV: {path}")
        # data.to_csv(path, index=False)

class JsonWriter(DataWriter):
    def write(self, data: pd.DataFrame, path: str):
        print(f"Đang ghi {len(data)} dòng ra file JSON: {path}")
        # data.to_json(path, orient='records')
```

**Câu hỏi 2:**
```python
class WriterFactory:
    def get_writer(self, format_type: str) -> DataWriter:
        if format_type == 'csv':
            return CsvWriter()
        elif format_type == 'json':
            return JsonWriter()
        else:
            raise ValueError(f"Định dạng không được hỗ trợ: {format_type}")

factory = WriterFactory()
csv_writer = factory.get_writer('csv')
json_writer = factory.get_writer('json')

print(csv_writer) # <__main__.CsvWriter object at ...>
print(json_writer) # <__main__.JsonWriter object at ...>
```

**Câu hỏi 3:**
```python
class FilterStrategy(ABC):
    @abstractmethod
    def filter(self, data: pd.DataFrame) -> pd.DataFrame:
        pass

class FilterByValue(FilterStrategy):
    def __init__(self, column: str, threshold: float):
        self.column = column
        self.threshold = threshold

    def filter(self, data: pd.DataFrame) -> pd.DataFrame:
        return data[data[self.column] > self.threshold]

class FilterByCategory(FilterStrategy):
    def __init__(self, column: str, category):
        self.column = column
        self.category = category

    def filter(self, data: pd.DataFrame) -> pd.DataFrame:
        return data[data[self.column] == self.category]
```

**Câu hỏi 4:**
```python
class ReportGenerator:
    def __init__(self, writer: DataWriter, filter_strategy: FilterStrategy):
        self._writer = writer
        self._filter_strategy = filter_strategy

    def generate(self, data: pd.DataFrame, path: str):
        print(f"Áp dụng chiến lược lọc: {self._filter_strategy.__class__.__name__}")
        filtered_data = self._filter_strategy.filter(data)
        print(f"Số dòng sau khi lọc: {len(filtered_data)}")
        self._writer.write(filtered_data, path)
        print("Tạo báo cáo hoàn tất.")
```

**Kịch bản 2 (`DatabaseConnection`) là ứng cử viên tốt nhất.**

**Giải thích:**
-   **Singleton phù hợp:** Việc quản lý kết nối cơ sở dữ liệu là một tài nguyên toàn cục và tốn kém. Việc tạo ra nhiều đối tượng quản lý kết nối riêng biệt sẽ rất lãng phí (mở nhiều kết nối không cần thiết) và có thể gây ra các vấn đề về đồng bộ. Do đó, việc đảm bảo chỉ có **một và chỉ một** đối tượng quản lý pool kết nối trong toàn bộ ứng dụng là một yêu cầu hợp lý. Singleton Pattern giải quyết chính xác vấn đề này.
-   **Singleton không phù hợp:**
    -   `User`: Một hệ thống phải có khả năng xử lý nhiều người dùng, mỗi người dùng có thông tin và trạng thái riêng. Mỗi `User` object phải là một instance riêng biệt. Áp dụng Singleton ở đây là sai hoàn toàn.
    -   `Product`: Tương tự, một trang web có hàng ngàn sản phẩm. Mỗi `Product` object phải đại diện cho một sản phẩm riêng biệt với các thuộc tính khác nhau (tên, giá, mô tả...). Singleton sẽ khiến bạn chỉ có thể có một sản phẩm duy nhất trong toàn bộ hệ thống.

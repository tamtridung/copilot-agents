```python
# Class Book đầy đủ từ câu 4
class Book:
    def __init__(self, title, author, page_count):
        self._title = title
        self._author = author
        self.page_count = page_count
    @property
    def title(self): return self._title
    @property
    def author(self): return self._author
    @property
    def page_count(self): return self._page_count
    @page_count.setter
    def page_count(self, value):
        if not isinstance(value, int) or value < 0:
            raise ValueError("Số trang phải là một số nguyên không âm.")
        self._page_count = value
    def __str__(self):
        return f"'{self.title}' của {self.author}"
    def __repr__(self):
        return f"Book(title='{self.title}', author='{self.author}', page_count={self.page_count})"

# Class BookCollection đầy đủ
class BookCollection:
    def __init__(self):
        self._books = []

    def add_book(self, book):
        if not isinstance(book, Book):
            raise TypeError("Chỉ có thể thêm object từ class Book vào bộ sưu tập.")
        self._books.append(book)
        print(f"Đã thêm cuốn '{book.title}' vào bộ sưu tập.")

    def __len__(self):
        return len(self._books)

    def __str__(self):
        return f"Bộ sưu tập với {len(self)} cuốn sách."

    def __getitem__(self, index):
        return self._books[index]

# Test code
```
